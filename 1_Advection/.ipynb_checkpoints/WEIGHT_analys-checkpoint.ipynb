```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
folder_to_read='data/output/weight/'
```

```python
files = os.listdir(folder_to_read)
weights=[i for i in files if 'weigth' in i]
biases=[i for i in files if 'bias' in i]
print(weights)
print(biases)
```

```python
weight_lst=[]
for weight in weights:
    weight_lst.append(pd.read_csv(folder_to_read+weight,index_col=0))
```

```python
bias_lst=[]
for bias in biases:
    bias_lst.append(pd.read_csv(folder_to_read+bias,index_col=0))
```

## **Weights**

```python
for weight_layer in weight_lst:
    print(weight_layer.shape)
    display(weight_layer)
    
    
    plt.figure(figsize=(10,8))
    sns.heatmap(weight_layer,cmap="coolwarm")
    plt.show()
```

## **Bias**

```python
for bias_layer in bias_lst:
    print(bias_layer.shape)
    display(bias_layer)
```
